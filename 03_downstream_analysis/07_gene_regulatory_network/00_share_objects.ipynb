## Spectra factor markers

```python
gene_factor_df = pd.read_csv(here('03_downstream_analysis/05_SPECTRA/results/SPECTRAFactor_Processed_LAM0.001_S0.6_NonCellIdentitymarkersDF.csv'), index_col=0)
gene_factor_df.columns = gene_weights.index
gene_factor_df
```

# 1. Add gene_id and gene_names in .var

```python
varDF = pd.read_csv(here('03_downstream_analysis/05_SPECTRA/results/varDF.csv'), index_col=0)
varDF
```

```python
adata.var = adata.var.merge(varDF, left_index=True, right_index=True, how='left')
adata.var
```

```python
adata.uns["Gene_Weights"]
```

# 5. Remove non relevant obs field

# 7.Include parameters

```python
adata.uns["Gene_selection_parameters"] = dict({"S": 0.6, 
                                              "LAM": 0.001})
```

# 8.Include selected genes in .var

```python
adata.var.drop(['mt', 'ribo', 'hb', 'plt'], axis=1, inplace=True)
```

```python
adata.var["SPECTRA_selected_genes"] = [True if g in gene_factor_df.index.tolist() else False for g in adata.var.index]
adata.var["SPECTRA_selected_genes"].value_counts()
```

```python
adata
```

# 8. Include initial signatures in .uns

```python
# Spectra signatures
with open(here('03_downstream_analysis/05_SPECTRA/results/SPECTRA_GeneSet_ensg_v2.pickle'), 'rb') as f:
    spectra_gene_sets = pickle.load(f)
```

```python
# V2 signatures
celltype = "B"
v2_signatures_dict = {}
for key in spectra_gene_sets[celltype]:
    if f"{celltype}_" in key:
        new_key = key.replace(f"{celltype}_", "")
        v2_signatures_dict[new_key] = spectra_gene_sets[celltype][key]
```

```python
v2_signatures_dict.keys()
```

```python
adata.uns["Input_signatures"] = v2_signatures_dict
```
